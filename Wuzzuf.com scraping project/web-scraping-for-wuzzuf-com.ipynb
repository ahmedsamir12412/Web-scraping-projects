### 4.1 Importing our Libraries & Methods for web scraping

```python
# BeautifulSoup is a Python library for pulling data out of HTML and XML files
from bs4 import BeautifulSoup as bs

# urllib.request for opening and reading URLs
from urllib.request import urlopen
```

```python
url='https://wuzzuf.net/search/jobs/?q=data+scientist'
```

### 4.2 Create a Client-based Request to Get the URL

```python
client = urlopen(url)
```

### 4.3 Getting the HTML Code of the Full Page


```python
html = client.read()
```

### 4.4 Closing the Request

```python
client.close()
```

### 4.5 Creating an HTML Parser Using BeautifulSoup

```python
soup = bs(html, "html.parser")
```

### 4.6 Creating a Container for the Needed Data

```python
containers = soup.find_all("div", {"class" : "css-1gatmva e1v1l3u10"})
```

```python
len(containers)
```

```python
bs.prettify(containers[0])
```

### 4.7 Accessing our desired info ( Job title , company name , Job type , Address .. )

```python
Job_title = containers[0].findAll("h2", {"class": "css-m604qf"})
print(job_title[0].text)
Company_name = containers[0].findAll("a", {"class": "css-17s97q8"})
print(company_name[0].text)
Job_type = containers[0].findAll("a", {"class": "css-n2jc4m"})
print(Job_type[0].text)
Address = containers[0].findAll("span", {"class": "css-5wys0k"})
print(Address[0].text)
```

### 4.8 Bringing it All Together

```python
# We need to create a file to save our new data to it
data_science_job_data = open('data_science_job_data.csv','w')
headers = "Job_title,Company_Name ,Job_type,Address\n"
data_science_job_data.write(headers)
```

```python
# Now we will get ALL the needed data from the web page
for container in containers:
    jtitle = container.findAll("h2", {"class": "css-m604qf"})
    Job_title = jtitle[0].text.strip()
    
    
    cname = container.findAll("a", {"class": "css-17s97q8"})
    Company_name = cname[0].text.strip()
    

    jtype = container.findAll("a", {"class": "css-n2jc4m"})
    Job_type = jtype[0].text.strip()
    
    adrs=container.findAll("span", {"class": "css-5wys0k"})
    Address=adrs[0].text.strip()
    
    
    
 
    
    print(Job_title + ", " + Company_name + ", " + Job_type + "," + Address)
```
